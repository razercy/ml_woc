```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


```

```python
df=pd.read_csv('unsupervised_data.csv')
df.head(15)
```

```python
X_train=df.iloc[:,1:].to_numpy()
print(X_train)
```

```python
centroids=X_train[:40000:8000]
print(centroids)

```

```python
for i in range(1000):
    b=np.argmin(np.sum((X_train[:,None,:]-centroids)**2,axis=2),axis=1)
    centroids=[np.mean(np.array([X_train[k] for k in range(40000) if b[k]==j]),axis=0) for j in range(5)]

    
```

```python
arr=np.argmin(np.sum((X_train[:,None,:]-centroids)**2,axis=2),axis=1)
print(arr)
```

```python
for i in range(5):
    print(np.count_nonzero(arr==i))
```

```python
print(np.array(centroids))
```

```python
print(np.sum((X_train-np.array(centroids)[arr])**2)/40000)
```
