Importing Packages: NumPy, Pandas, Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_train=pd.read_csv('linear_regression_train.csv')
df_train.head(15)
```

```python
X_train=df_train.iloc[:,1:-1].to_numpy()
print(X_train)
```

```python
y_train=df_train.iloc[:,-1].to_numpy()
print(y_train)
```

```python
X_train_norm=X_train.copy()
mu=np.mean(X_train_norm,axis=0)
sigma=np.std(X_train_norm,axis=0)
X_train_norm=(X_train_norm-mu)/sigma
print(X_train_norm)
```

```python
print(y_train)
```

```python
w=np.zeros_like(df_train.columns[1:-1])
print(w)
```

```python
b=np.mean(y_train)
print(b)
```

```python
J_hist=[]
for i in range(1000):
    if i%100==0:
        errors=np.dot(X_train_norm,w)+b-y_train
        J_hist.append(np.dot(errors.T,errors)/(2*X_train_norm.shape[0]))
        print(J_hist[-1])
    dj_dw=np.dot(X_train_norm.T,np.dot(X_train_norm,w)+b-y_train)/X_train_norm.shape[0]
    dj_db=np.dot(np.ones_like(X_train_norm[:,0]),(np.dot(X_train_norm,w)+b-y_train))/X_train_norm.shape[0]
    w-=1*dj_dw
    b-=1*dj_db
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,constrained_layout=True,figsize=(12, 4))
ax1.plot(J_hist)
ax2.plot(100 + np.arange(len(J_hist[100:])),J_hist[100:])
ax1.set_title("Cost vs. iteration");  ax2.set_title("Cost vs. iteration (tail)")
ax1.set_ylabel('Cost')             ;  ax2.set_ylabel('Cost') 
ax1.set_xlabel('iteration step')   ;  ax2.set_xlabel('iteration step') 
plt.show()
```
